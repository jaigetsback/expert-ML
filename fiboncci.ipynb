```python
import matplotlib.pyplot as plt

# Function to calculate Fibonacci numbers
def fibonacci(n):
    fib = [0, 1]
    for i in range(2, n):
        fib.append(fib[i-1] + fib[i-2])
    return fib

# Calculate the first 100 Fibonacci numbers
fib_numbers = fibonacci(100)
print (fib_numbers)

# Create a plot
plt.figure(figsize=(10, 6))
plt.plot(range(1, 101), fib_numbers, marker='o', linestyle='-')
plt.title('First 100 Fibonacci Numbers')
plt.xlabel('Index')
plt.ylabel('Fibonacci Number')
plt.grid(True)
plt.show()
```

```python
import sys
## This part of the code is to be run from the command line with sys.argv call 
#print("Distance in Km:", sys.argv[1])
#print("Time in hours:", sys.argv[2])
#distance_in_km = int(sys.argv[1])
#time_in_hours = int(sys.argv[2])
distance_in_km = float(input("Enter distance in km: "))
time_in_hours  = float(input("Enter time in hours: "))

distance_in_mi = distance_in_km / 1.6
distance_in_mtrs = distance_in_km * 1000

time_in_seconds = time_in_hours * 3600

speed_in_kph = distance_in_km / time_in_hours
speed_in_mph = distance_in_mi / time_in_hours
speed_in_mps = distance_in_mtrs / time_in_seconds

print("The speed in kilometers per hour is:", speed_in_kph)
print("The speed in miles per hour is:", speed_in_mph)
print("The speed in meters per second is:", speed_in_mps)
```
